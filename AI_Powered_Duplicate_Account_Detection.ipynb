```python
import pandas as pd

data = {
    "customer_id":[101,102,103,104,105],
    "name":["Ravi Kumar","R. Kumar","Arjun Reddy","Rahul Sharma","Ravi Kumar"],
    "phone":["9876543210","9876543210","9123456789","9988776655","9876543210"],
    "email":["ravi@gmail.com","ravi.k@gmail.com","arjun@gmail.com",
             "rahul@gmail.com","ravi@gmail.com"],
    "address":["Hyderabad","Hyderabad","Vizag","Delhi","Hyderabad"],
    "pan":["ABC1234X","ABC1234X","XYZ5678P","PQR9876Z","ABC1234X"]
}

df = pd.DataFrame(data)

df.to_csv("customers.csv",index=False)

df
```

```python
from rapidfuzz import fuzz
import pandas as pd

df = pd.read_csv("customers.csv")

duplicates = []

for i in range(len(df)):
    for j in range(i+1,len(df)):

        name_score = fuzz.ratio(df.loc[i,'name'],df.loc[j,'name'])

        phone_match = df.loc[i,'phone']==df.loc[j,'phone']
        pan_match = df.loc[i,'pan']==df.loc[j,'pan']

        if (name_score > 70 and phone_match) or pan_match:
            duplicates.append({
                "Customer1":df.loc[i,'customer_id'],
                "Customer2":df.loc[j,'customer_id'],
                "Name Similarity":name_score,
                "Phone Match":phone_match,
                "PAN Match":pan_match
            })

duplicate_df = pd.DataFrame(duplicates)

duplicate_df
```

```python
def risk_score(row):
    score = 0

    if row["Phone Match"]:
        score += 40

    if row["PAN Match"]:
        score += 40

    if row["Name Similarity"] > 80:
        score += 20

    return score

duplicate_df["Risk Score"] = duplicate_df.apply(risk_score,axis=1)

duplicate_df
```

```python
import pandas as pd
from rapidfuzz import fuzz

df = pd.read_csv("customers.csv")

duplicates = []

for i in range(len(df)):
    for j in range(i+1, len(df)):

        name_score = fuzz.ratio(df.loc[i,'name'], df.loc[j,'name'])
        phone_match = df.loc[i,'phone'] == df.loc[j,'phone']
        pan_match = df.loc[i,'pan'] == df.loc[j,'pan']

        if (name_score > 70 and phone_match) or pan_match:
            duplicates.append({
                "Customer1": df.loc[i,'customer_id'],
                "Customer2": df.loc[j,'customer_id'],
                "Name Similarity": name_score,
                "Phone Match": phone_match,
                "PAN Match": pan_match
            })

def risk_level(score):
    if score >= 90:
        return "HIGH"
    elif score >= 60:
        return "MEDIUM"
    else:
        return "LOW"

duplicate_df["Risk Level"] = duplicate_df["Risk Score"].apply(risk_level)

print(duplicate_df)
```
